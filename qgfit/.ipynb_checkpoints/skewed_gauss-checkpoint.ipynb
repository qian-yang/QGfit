```python
import numpy as np
import units

def cal_velocity(d_ln_lambda):
    """
    Return velocity in units of km/s
    """
    # velocity = Delta_lambda / lambda * C = d_ln_lambda * C
    velocity = d_ln_lambda * units.C_KMS
    return velocity

def voigt_fwhm(sigma, gamma):
    """
    Return FWHM in units of km/s
    The input parameters was fitted in log scale
    """
    # The lines are in log guass
    # FWHM = Delta_lambda / lambda * C
    # Delta_lambda / lambda = Delta(log_lambda) = width in log lambda
    width_gauss = 2.3548 * sigma
    width_lorentz = 2.0 * gamma
    width_voigt = 0.5346 * width_lorentz + np.sqrt(0.2166 * width_lorentz**2 + width_gauss**2)
    fwhm_velocity = cal_velocity(width_voigt)
    return fwhm_velocity

log_sigma = 0.005
gamma = log_sigma * 1.0
print(voigt_fwhm(log_sigma, gamma))

# gauss
log_sigma = 0.002
gamma = 0
print(voigt_fwhm(log_sigma, gamma))

log_sigma = 0.0017
print(voigt_fwhm(log_sigma, gamma))

log_sigma = 2.3E-4
print(voigt_fwhm(log_sigma, gamma))

log_sigma = 0.05
print(voigt_fwhm(log_sigma, gamma))

log_sigma = 0.001
sigma = log_sigma
print(voigt_fwhm(log_sigma, gamma))
print((0.5346 * (2.0 * gamma) + np.sqrt(0.2166 * (2.0 * gamma)**2 + (2.3548 * sigma)**2)) * 2.99792458e5)

ln_sigma = log_sigma
print((0.5346 * (2.0 * gamma * ln_sigma) + np.sqrt(0.2166 * (2.0 * gamma * ln_sigma)**2 + (2.3548 * ln_sigma)**2)) * 2.99792458e5)

print('----- skew -----')
log_sigma = 0.0037  # 3994.667166622189 initial for broad emission
print(voigt_fwhm(log_sigma, log_sigma))

log_sigma = 0.0017 
print(voigt_fwhm(log_sigma, log_sigma))  # 1835.3876170966814
print(voigt_fwhm(log_sigma, 0)) # 1200.1171761672797

log_sigma = 0.001  # min for broad emission
print(voigt_fwhm(log_sigma, log_sigma))  # 1079.639774762754
print(voigt_fwhm(log_sigma, 0)) # 705.9512800984

# min (gamma_rato max = 1e5)
log_sigma = 1.7e-8  # real min for broad emission
print(voigt_fwhm(log_sigma, log_sigma * 1e5))  # 1019.2974673501784  # allow a full range of log_sigma
print(voigt_fwhm(log_sigma, log_sigma))  # 0.018353876170966815

log_sigma = 2.7e-9  # real min for narrow emission
print(voigt_fwhm(log_sigma, log_sigma * 1e5))  # 161.88842128502833

# max (gamma_ratio = 0, gauss)
log_sigma = 0.0017  # real max for narrow emission
print(voigt_fwhm(log_sigma, 0))  # 1200.1171761672797, allow a full range of gamma_ratio
print(voigt_fwhm(log_sigma, log_sigma))  # 1835.3876170966814

log_sigma = 0.029  # real max for broad emission
print(voigt_fwhm(log_sigma, 0))  # 20472.5871228536
print(voigt_fwhm(log_sigma, log_sigma))  # 31309.55346811987

# init gamma_ratio = 1
log_sigma = 0.00047  # real init for narrow emission
print(voigt_fwhm(log_sigma, log_sigma))  # 507.4306941384944

log_sigma = 0.0037  # real init for broad emission
print(voigt_fwhm(log_sigma, log_sigma))  # 3994.667166622189

# voff broad, 0 - 4000
print(cal_velocity(0.015))  # 4496.886869999999

# voff narrow, 0 - 4000
print(cal_velocity(0.005))  # 1498.96229
```

```python
# init gauss for narrow

import numpy as np
import units

def cal_velocity(d_ln_lambda):
    """
    Return velocity in units of km/s
    """
    # velocity = Delta_lambda / lambda * C = d_ln_lambda * C
    velocity = d_ln_lambda * units.C_KMS
    return velocity

def voigt_fwhm(sigma, gamma):
    """
    Return FWHM in units of km/s
    The input parameters was fitted in log scale
    """
    # The lines are in log guass
    # FWHM = Delta_lambda / lambda * C
    # Delta_lambda / lambda = Delta(log_lambda) = width in log lambda
    width_gauss = 2.3548 * sigma
    width_lorentz = 2.0 * gamma
    width_voigt = 0.5346 * width_lorentz + np.sqrt(0.2166 * width_lorentz**2 + width_gauss**2)
    fwhm_velocity = cal_velocity(width_voigt)
    return fwhm_velocity

# min (gamma_rato max = 1 for narrow)
log_sigma = 0.00015  # real min for narrow emission
print(voigt_fwhm(log_sigma, log_sigma))  # 161.94596621441306
print(voigt_fwhm(log_sigma, 0))  # 105.89269201475999

# # max (gamma_ratio = 0, gauss)
log_sigma = 0.0017  # real max for narrow emission
print(voigt_fwhm(log_sigma, 0))  # 1200.1171761672797, allow a full range of gamma_ratio
print(voigt_fwhm(log_sigma, log_sigma))  # 1835.3876170966814

# init gamma_ratio = 1
log_sigma = 0.0005  # real init for narrow emission
print(voigt_fwhm(log_sigma, 0))  # 352.9756400492
print(voigt_fwhm(log_sigma, log_sigma))  # 539.819887381377

# setting combination for narrow
log_sigma = 0.0005
print(voigt_fwhm(log_sigma, 1e-8 * log_sigma))  # 352.97564165189044
print(voigt_fwhm(log_sigma, 1e-3 * log_sigma))  # 353.13593667285573
print(voigt_fwhm(log_sigma, 1e-2 * log_sigma))  # 354.58108807989555
print(voigt_fwhm(log_sigma, 1e-1 * log_sigma))  # 369.27819332300487

# setting combination for broad
log_sigma = 0.0037
print(voigt_fwhm(log_sigma, 1.0 * log_sigma))  # 3994.667166622189


# in practice
#       linename      norm  ln_sigma  gamma_ratio      skew          FWHM
# 0    Halpha_br  0.024815  0.001330     3.718482  0.003285   3254.267518
# 1    Halpha_na  0.002922  0.000739     0.011029  0.000000    524.359195

log_sigma = 0.000739
print(voigt_fwhm(log_sigma, 0.011029 * log_sigma))  # 524.3154772082676

log_sigma = 0.001330
print(voigt_fwhm(log_sigma, 3.718482 * log_sigma))  # 3254.414909907126
```
