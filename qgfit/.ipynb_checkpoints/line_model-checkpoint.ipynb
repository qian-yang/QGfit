```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from model import conti_model_all_config, host_sed_all_out
from line import many_gauss_off, com_cols, cal_FWHM
from write import read_data

def get_line_flux(xval, lines):
    pp_line = com_cols(lines['norm'], lines['ln_shift'], lines['ln_sigma'])
    ln_lambda = np.array(lines['ln_lambda'])
    flux_line = many_gauss_off(xval, pp_line, ln_lambda)
    return flux_line

def get_ind(ind, *args):
    args = (arg[ind] for arg in args)
    return args

file = 'J1011+5442_one.pkl'
obj = read_data(file)

conti = obj['conti']
line = obj['line']
pars_set = conti['pars_set']

data = obj['data_rest'][0]
print(data)
wave, flux, fluxerr = data

pars_one = pars_set[0]
flux_conti, flux_pl, flux_fe_uv, flux_fe_op, flux_balmer, \
                flux_host = conti_model_all_config(wave,
                pars_one, obj['config'])
flux_resi = flux - flux_conti
# flux_line = get_line_flux(xval, lines)

ind = (wave > 6400) & (wave < 6800)
wave, flux, fluxerr, flux_resi = get_ind(ind, wave, flux, fluxerr, flux_resi)

data_line = (wave, flux_resi, fluxerr)

import pandas as pd
from lmfit import Parameters, minimize, report_fit
import timeit

file = 'template/qsopar_broad_sep_voff.csv'
LINE_LIST = pd.read_csv(file)
lines = LINE_LIST

def create_lines(line_list):
    """
    create repeat rows when n_gauss > 1
    """
    lines = None
    for k in range(len(line_list)):
        line = line_list[k: k + 1]
        n_gauss = int(line['ngauss'])
        if n_gauss > 1:
            line = pd.DataFrame(np.repeat(line.values, n_gauss,
                axis=0), columns=line.columns)
        if lines is None:
            lines = line
        else:
            lines = pd.concat([lines, line], ignore_index=True)
    lines['ln_lambda'] = np.log(lines['lambda'].astype('float'))
    # reset index
    lines.reset_index(drop=True, inplace=True)
    return lines

def init_line_par(lines):
    # --- key function that is different from kmpfit ---
    parf = Parameters()
    for k_tp, line in lines.iterrows():
        suffix = '_' + str(k_tp)
        off = np.abs(line['voff'])
        parf.add('norm' + suffix, 0.01, min=0.0, max=1e10)
        parf.add('ln_shift' + suffix, 0.0, min=-off, max=+off)
        parf.add('ln_sigma' + suffix, line['inisig'],
            min=line['minsig'],
            max=line['maxsig'])
    parf = tie_line(parf, lines)
    return parf

def tie_line(parf, lines):
    tie_width = True
    tie_shift = True
    tie_flux = True
    # --- from lines_new ---
    if tie_width:
        parf = tie_par(parf, 'ln_sigma', lines['windex'])
    # ---
    if tie_shift:
        parf = tie_par(parf, 'ln_shift', lines['vindex'])
    # ---
    if tie_flux:
        parf = tie_par(parf, 'norm', lines['findex'],
            tie_fvalue=lines['fvalue'])
    return parf

def tie_par(parf, key, tie, tie_fvalue=None):
    """
    For lmfit
    key = 'norm', 'ln_shift', or 'ln_sigma'
    tie = lines['windex'], or lines['vindex'], lines['findex']
    tie_fvalue = lines['fvalue'], or default as None.
    tie_factor = tie_fvalue, or default as 1.
    """
    # NOTE: tie_value = 1 or 2, and etc.
    tie_value, ind_uniq = np.unique(tie, return_index=True)
    nt = len(tie_value)
    for value in tie_value:
        if value > 0:
            index_tie = np.where(tie == value)[0]
            ind0 = index_tie[0]
            suffix0 = '_' + str(ind0)
            for ind in index_tie:
                if ind != ind0:
                    if tie_fvalue is None:
                        tie_factor = 1.0
                    else:
                        tie_factor = tie_fvalue[ind] / tie_fvalue[ind0]
                    # ---
                    suffix = '_' + str(ind)
                    parf[key + suffix].set(expr='{0}{1} * {2}'.format(key, suffix0, tie_factor))
    return parf

def time_interval(start, stop, num=1):
    dt = np.round(stop - start, num)
    dt_str = str(dt)
    return dt_str

def my_minimize(params, data_line, comp_name, method='leastsq',
              **kwargs):
    start = timeit.default_timer()

    out = minimize(residual_line, params,
        args=(data_line,),
        method=method,
        nan_policy='omit', **kwargs)

    stop = timeit.default_timer()
    print('Line: {0} {1} {2}s, redchi2 = {3}'.format(comp_name, method,
        time_interval(start, stop), str(np.round(out.redchi, 2))))
    return out
    # todo: debug when redchi is large, emcee fails
    

def get_par(parf, type='value'):
    """
    create p_value array in order of [norm, ln_shift, ln_sigma]
    """
    keys = ['norm', 'ln_shift', 'ln_sigma']

    ngauss = int(len(parf) / 3.0)
    par_value = []
    for k in range(ngauss):
        suffix = '_' + str(k)
        for key in keys:
            par_value.append(getattr(parf[key + suffix], type))
    return par_value
    
def residual_line(parf, data_line):
    wave, flux, fluxerr = data_line
    par_value = get_par(parf)
    line_model = many_gauss_off(wave, par_value, ln_lambda_temp)
    residual_line = (flux - line_model) / fluxerr
    residual_line = np.asarray(residual_line, dtype=np.float64)
    return residual_line

def sep_cols(array):
    """separate columns"""
    array = np.array(array)
    n_gauss = int(len(array) / 3)
    ind = np.arange(n_gauss) * 3
    col1, col2, col3 = array[ind], array[ind + 1], array[ind + 2]
    return (col1, col2, col3)

def update_res(lines, line_fit, suffix=''):
    par_value = get_par(line_fit)
    std_value = get_par(line_fit, type='stderr')
    # --- add columns ---
    lines['norm' + suffix], lines['ln_shift' + suffix], \
        lines['ln_sigma' + suffix] = sep_cols(par_value)
    # --- stderr ---
    lines['norm_stderr' + suffix], lines['ln_shift_stderr' + suffix], \
        lines['ln_sigma_stderr' + suffix] = sep_cols(std_value)
    return lines

comp_name = 'Ha'
index_line = np.where(lines['compname'] == comp_name)[0]
lines_comp = lines.iloc[index_line]

lines_new = create_lines(lines_comp)
parf = init_line_par(lines_new)

lines_temp = lines_new
ln_lambda_temp = np.array(lines_new['ln_lambda'])

out = my_minimize(parf, data_line, comp_name, method='leastsq')

lines = update_res(lines_temp, out.params)
xval = np.log(wave)
flux_line = get_line_flux(xval, lines)
# print(lines)

# report_fit(out)
n_data = len(wave)

plt.figure(figsize=(15, 8))
plt.step(wave, flux_resi, color='k')
plt.plot(wave, flux_line, color='r')
# print(len(wave))
```
